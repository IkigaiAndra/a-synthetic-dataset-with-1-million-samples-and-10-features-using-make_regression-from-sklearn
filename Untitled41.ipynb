```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import time

# Generate a large synthetic dataset
n_samples = 10**6  # 1 million samples
n_features = 10    # 10 features

X, y = make_regression(n_samples=n_samples, n_features=n_features, noise=0.1)

# Initialize parameters
learning_rate = 0.01
n_iterations = 1000
batch_size = 1000

# Initialize weights and bias for linear regression
weights = np.random.randn(n_features)
bias = np.random.randn()

# Helper function for mini-batch gradient descent
def sgd_step(X_batch, y_batch, weights, bias, learning_rate):
    # Predictions
    predictions = np.dot(X_batch, weights) + bias

    # Compute the gradients
    error = predictions - y_batch
    grad_w = (2 / len(X_batch)) * np.dot(X_batch.T, error)
    grad_b = (2 / len(X_batch)) * np.sum(error)

    # Update weights and bias
    weights -= learning_rate * grad_w
    bias -= learning_rate * grad_b

    return weights, bias

# Mini-batch Stochastic Gradient Descent
start_time = time.time()

for iteration in range(n_iterations):
    # Shuffle the dataset to create mini-batches
    shuffle_indices = np.random.permutation(len(X))
    X_shuffled = X[shuffle_indices]
    y_shuffled = y[shuffle_indices]

    # Process mini-batches
    for i in range(0, len(X), batch_size):
        X_batch = X_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]

        # Perform one SGD step
        weights, bias = sgd_step(X_batch, y_batch, weights, bias, learning_rate)

    # Optionally, print progress
    if iteration % 100 == 0:
        print(f"Iteration {iteration + 1}/{n_iterations} - Time elapsed: {time.time() - start_time:.2f}s")

# Final weights and bias
print(f"Final weights: {weights}")
print(f"Final bias: {bias}")

print(f"Training completed in {time.time() - start_time:.2f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Generate large dataset
x = np.linspace(0, 10, 100000)  # 100,000 data points
y = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)  # Sinusoidal data with noise

# Apply cubic spline interpolation
cs = CubicSpline(x, y)

# Create new set of points for smoother plotting
x_new = np.linspace(0, 10, 1000)  # 1000 points for smoother curve
y_new = cs(x_new)

# Plot the original data and the interpolated data
plt.figure(figsize=(10, 6))
plt.plot(x, y, '.', markersize=1, label="Noisy Data")
plt.plot(x_new, y_new, '-', label="Cubic Spline Interpolation", linewidth=2)
plt.title("HorseHedge Method: Interpolation on Large Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
